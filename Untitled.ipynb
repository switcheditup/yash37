```python
import pandas as pd
```

```python
f=pd.read_csv("Employee.csv")
```

```python
f.head()
```

```python
#identifying values
print(f.isnull().sum())
```

```python
#then handle the missing values
f['Salary'] = pd.to_numeric(f['Salary'], errors='coerce')
f['Salary'].fillna(f['Salary'].mean(),inplace=True)
```

```python
f.head()
```

```python
f.drop(columns=['Id'],inplace=True)
```

```python
f.head()
```

```python
#detect outliers using IQR
#formula Q3 - Q1 
#outliers

Q1=f['Salary'].quantile(0.25)

Q3=f['Salary'].quantile(0.75)

IQR= Q3 - Q1
```

```python
outliners = f[(f['Salary'] < Q1 - 1.5 * IQR) | (f['Salary'] > Q3 + 1.5 * IQR)]
```

```python
print(outliners)
```
